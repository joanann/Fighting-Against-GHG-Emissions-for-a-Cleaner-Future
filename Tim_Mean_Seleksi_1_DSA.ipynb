```python
#Outliers
h = sns.catplot(x='Tanggal', y="Dirawat Harian (Indonesia)", data=dataindojak)
```

```python
df1 = dataindojak[["Tanggal","Dirawat Harian (Jakarta)"]]
df1.head()
```

```python
df1["Tanggal"] = pd.to_datetime(df1["Tanggal"])
df = df1.set_index("Tanggal")
df["year"] = df.index.year
df["month"] = df.index.month
df["day"] = df.index.day
df.head()
```

```python
x = df.groupby(["year", "month", "day"])["Dirawat Harian (Jakarta)"].mean()
df_wide = x.unstack()
df_wide.head()
```

```python
sns.boxplot(x = 'month', y='Dirawat Harian (Jakarta)', data = df)
```

```python
sns.boxplot(x = 'year', y='Dirawat Harian (Jakarta)', data = df)
```

```python
df2 = dataindojak[["Tanggal","Positif Aktif Harian (Jakarta)"]]
df2.head()
```
